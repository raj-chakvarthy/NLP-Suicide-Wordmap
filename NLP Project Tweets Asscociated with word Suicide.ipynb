>Using searchTwitter function from TwitteR library to search for word Suicide, n= no of tweets to analyse , lang= languge of tweets to analyse

```r
depressed.tweets<- searchTwitter('suicide',n=5000,lang='en')
```

```r
text_tweets<-sapply(depressed.tweets,function(x) x$getText()) #extracting the words
```

```r
text_tweets<- iconv(text_tweets,'UTF-8','ASCII') #removing 
```

```r
text_corpus<-Corpus(VectorSource(text_tweets)) # creating a corpus of extracted words
```

>Creating a Term Doc Matrix, stopwords= vector containg words which will be excluded from wordmap. Ex: Suicide is excluded as we are intrested in words contained in the tweets which has suicide in it, not the actual word suicide

```r
term.doc.matrix<- TermDocumentMatrix(text_corpus,control=list(removePunctuation=TRUE,stopwords=c('suicide','https','the','you','your','for','being','and','this','that'),removeNumbers=TRUE,tolower=TRUE))
```

```r
term.doc.matrix <-as.matrix(term.doc.matrix)
```

```r
word.freq <- sort(rowSums(term.doc.matrix),decreasing = TRUE)
dm<- data.frame(word=names(word.freq),freq=word.freq)
```

<p><h3>Wordmap of words associated with Suicide</h3></p>

```r
wordcloud(dm$word,dm$freq,random.order = FALSE,colors=brewer.pal(8,'Dark2'))
```
